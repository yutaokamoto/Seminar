```python
%%cython
import numpy as np

#ok
cdef dict E_dict_func(list E): #キーにアイテム名、値にそのアイテムとコンフリクトのあるアイテムをもつ辞書
    cdef:
        dict E_dict = {}
    for e in E:
        E_dict[e[0]] = tuple(set(E_dict.setdefault(e[0],()) + (e[1],)))
        E_dict[e[1]] = tuple(set(E_dict.setdefault(e[1],()) + (e[0],)))
    return E_dict


#ok
cdef list nonconf_E_func(list E,list I): #枝集合　アイテムiとjにコンフリクトがないとき、枝(i,j)をつくる
    cdef:
        list nonconf_E = [(i,j) for i in I for j in I if (i!=j and {(i,j),(j,i)}&set(E)==set())]
    return nonconf_E


#ok
cdef list update_E(list E,list I): #アイテムの分割の際、コンフリクトがあるアイテム間の枝の集合を更新する
    cdef:
        list E_prime = [e for e in E if ( ({e[0],e[1]}&set(I))!=set() )]
    return E_prime


#ok
cdef list update_M(list M,list I):
    cdef:
        list M_prime
    M_prime = list(tuple(set(I)&set(g)) for g in M if len(set(I)&set(g))!=0)
    return M_prime


#ok
cdef bint capacity(int w_i,list bin_b,list w,int C): #bin_bの容量がw_iの大きさよりも大きいかを判定
    if (C-sum(w[i] for i in bin_b)) >= w_i:
        return True
    else:
        return False

    
#ok
cdef bint conflict(int item,list item_in_bin_b,dict E_dict): #bin_bに割り当てられているアイテムたちとw_iの間にコンフリクトがあるかを判定
    if item not in set(E_dict.keys()):
        return True
    for i in item_in_bin_b:
        if i in set(E_dict[item]):
                return False
    return True


#ok
cdef int group(int i,list M): #アイテムiの属するグループの番号を返す関数
    for index,menber in enumerate(M):
        if i in set(menber):
            return index


#ok
cdef bint degree(int i,dict E_dict): #アイテムiのコンフリクトの次数を返す関数
    try:
        return len(E_dict[i])
    except:
        return 0
    
    
#ok
cdef list sort_M(list M,dict E_dict,list r):
    if M!=[] and E_dict!={}:
        return sorted(M,key=lambda x:(r[M.index(x)], max(len(E_dict[i]) if i in set(E_dict.keys()) else 0 for i in x)),reverse=True)
    else:
        return []

    
#ok
cdef dict FFD(list I,int C,list w,list M,dict assign): #アイテムをビンに割り当てる
    for i in I:
        for b in range(len(assign.keys())+1): #インデックスが小さい順にビンを走査
            assign[b] = assign.setdefault(b,[])
            if capacity(w[i],assign[b],w,C): #容量制約を満たす
                assign[b].append(i) #ビンbにアイテムiを入れる
                break   
    return assign


#ok
cdef dict FFD_for_conflict(list I,int C,list w,dict E_dict,list M,dict assign): #コンフリクトの数が多い順にアイテムをビンに割り当てる
    for i in I:
        for b in range(len(assign.keys())+1): #インデックスが小さい順にビンを走査
            assign[b] = assign.setdefault(b,[])
            if capacity(w[i],assign[b],w,C) and conflict(i,assign[b],E_dict): #容量制約を満たす　and　コンフリクトがない
                assign[b].append(i) #ビンbにアイテムiを入れる
                break  
    return assign


#ok
cdef dict H3(list I,int C,list w,list E,list M,list r,dict assign): #コンフリクトの数が多い順にアイテムをビンに割り当てる、をグループごとに行う
    cdef:
        dict E_dict = {}
        list M_prime = []
    E_dict = E_dict_func(E) #コンフリクトを表す枝を管理する辞書
    M_prime = update_M(M,I) #グループの順序つき集合Mを更新する。
    M_prime = sort_M(M_prime,E_dict,r)
    for g in M_prime: #グループごとにFFD_for_conflict()
        assign = FFD_for_conflict(list(g),C,w,E_dict,M_prime,assign) #gに対してFFD for conflictを行う
    return assign


#ok
cdef dict H4(list I,int C,list w,list E,list M,list r,dict assign): #コンフリクトの数が多い順にアイテムをビンに割り当てる、をグループごとに行う
    cdef:
        dict E_dict = {}
        list M_prime = []
        list H = []
    E_dict = E_dict_func(E) #コンフリクトを表す枝を管理する辞書
    H = list(set(I)-set(E_dict.keys())) #アイテムの順序付き集合I内のコンフリクトの数が0のアイテムの集合をHとする
    I = list(set(I)-set(H)) #I=I\Hとする。
    I = sorted(I,key=lambda x:(degree(x,E_dict), w[x]),reverse=True) #アイテム集合をアイテムの大きさが大きい順に並べ替える
    M_prime = update_M(M,I) #グループの順序つき集合Mを更新する。
    M_prime = sort_M(M_prime,E_dict,r)
    for g in M_prime: #グループごとにFFD_for_conflict()
        assign = FFD_for_conflict(list(g),C,w,E_dict,M_prime,assign) #gに対してFFD for conflictを行う   
    assign = FFD(H,C,w,M_prime,assign) #Hに対してFFDを行う
    return assign


#ok
cdef dict H5(list I,int C,list w,list E,list M,list r,dict assign): #コンフリクトの数が多い順にアイテムをビンに割り当てる、をグループごとに行う
    cdef:
        dict E_dict = {}
        list M_prime = []
        list H = []
    E_dict = E_dict_func(E) #コンフリクトを表す枝を管理する辞書
    M_prime = update_M(M,I) #グループの順序つき集合Mを更新する。
    M_prime = sort_M(M_prime,E_dict,r)
    for g in M_prime: #グループごとにFFD_for_conflict()
        H = list(set(g)-set(E_dict.keys())) #アイテムの順序付き集合I内のコンフリクトの数が0のアイテムの集合をHとし、I=I\Hとする。
        g = set(g)-set(H) #g=g\H
        g = sorted(list(g),key=lambda x:w[x],reverse=True) #アイテム集合をアイテムの大きさが大きい順に並べ替える
        assign = FFD_for_conflict(g,C,w,E_dict,M_prime,assign) #gに対してFFD for conflictを行う
        assign = FFD(H,C,w,M_prime,assign) #Hに対してFFDを行う
    return assign


#ok
cdef dict H6(list I,int C,list w,list E,list M,list r,dict assign): #コンフリクトの数が多い順にアイテムをビンに割り当てる、をグループごとに行う
    cdef:
        dict E_dict = {}
        list M_prime = []
        list H = []
    E_dict = E_dict_func(E) #コンフリクトを表す枝を管理する辞書
    M_prime = update_M(M,I) #グループの順序つき集合Mを更新する。
    M_prime = sort_M(M_prime,E_dict,r)
    for g in M_prime:
        H = list(set(g)-set(E_dict.keys())) #アイテムの順序付き集合I内のコンフリクトの数が0のアイテムの集合をHとし、I=I\Hとする。
        g = tuple(set(g)-set(H)) #g=g\H
        while len(g)>0:
            max_degree = max(g,key=lambda x:len(E_dict[x])) #グループg内で最もコンフリクトの数が多いアイテムiを選ぶ
            g_i = list({max_degree}|({g}&( set(I)-set(E_dict[max_degree]) ) ) ) #グループg内でiとコンフリクトがないアイテムとiからなる集合g_iを定める
            assign = FFD_for_conflict(g_i,C,w,E_dict,M_prime,assign) #g_iに対してFFD for conflictを行う
            g = tuple(set(g)-set(g_i)) #g=g\g_i
        assign = FFD(H,C,w,M_prime,assign) #Hに対してFFDを行う
    return assign


#ok
cdef H7(list I,int C,list w,list E,list M,list r,dict assign): #コンフリクトの数が多い順にアイテムをビンに割り当てる、をグループごとに行う
    cdef:
        dict E_dict = {}
        list M_prime = []
        list H = []
    E_dict = E_dict_func(E) #コンフリクトを表す枝を管理する辞書
    H = list(set(I)-set(E_dict.keys())) #アイテムの順序付き集合I内のコンフリクトの数が0のアイテムの集合をHとし、I=I\Hとする。
    I = list(set(I)-set(H))
    I = sorted(I,key=lambda x:(degree(x,E_dict), w[x]),reverse=True) #アイテム集合をアイテムの大きさが大きい順に並べ替える
    M_prime = update_M(M,I)#[tuple(set(I)&set(g)) for g in M if len(set(I)&set(g))!=0] #グループの順序つき集合Mを更新する。
    M_prime = sort_M(M_prime,E_dict,r)
    for g in M_prime:
        while len(g)>0:
            max_degree = max(g,key=lambda x:len(E_dict[x])) #グループg内で最もコンフリクトの数が多いアイテムiを選ぶ
            g_i = list({max_degree}|({g}&( set(I)-set(E_dict[max_degree]) ) ) ) #グループg内でiとコンフリクトがないアイテムとiからなる集合g_iを定める
            assign = FFD_for_conflict(g_i,C,w,E_dict,M_prime,assign) #g_iに対してFFD for conflictを行う
            g = tuple(set(g)-set(g_i)) #g=g\g_i
    assign = FFD(H,C,w,M_prime,assign) #Hに対してFFDを行う
    return assign

cdef dict H8(list I,int C,list w,list E,list M,list r,dict assign):
    cdef:
        list M_prime = []
    M_prime = update_M(M,I)#[tuple(set(I)&set(g)) for g in M if len(set(I)&set(g))!=0] #グループの順序つき集合Mを更新する。
    M_prime = sort_M(M_prime,E_dict,r)
    for g in M_prime:
        assign = Gendreau(I,C,w,E,M,assign)
    return assign


#ok
cdef bint clique(set SUB,set A):
    cdef bint flag=True
    for i in SUB:
        for j in SUB:
            if i!=j:
                if (i,j) not in A and (j,i) not in A:
                    flag = False
                    break
        if flag!=True:
            break
    return flag


#ok
cdef set fewest_degree_node(set SUB,set A):
    cdef:
        dict degree_dict = {}
        set fewest_degree = set()
    for n in SUB:
        degree_dict[n] = 0
    for e in A:
        try:
            degree_dict[e[0]] += 1
            degree_dict[e[1]] += 1
        except:
            continue
    fewest_degree = set(min(degree_dict.items(),key=lambda x:x[1]))
    return fewest_degree


#ok
cdef set Johnson(set N,set A): #与えられた点集合と枝集合から最大クリークを生成する #Nは点の集合、Aは枝の集合
    cdef:
        set SUB = N
        set y =set()
    while True:
        if clique(SUB,A):
            break
        else:
            y = fewest_degree_node(SUB,A) #SUBのノードのうち、最も次数の少ないものを定める
            if y==SUB:
                break
            SUB = SUB-y #SUBからyを除く
            for e in A:
                if y in e:
                    A = A-{e} #yにつながってる枝を除く
    return SUB


#ok
cdef dict Gendreau(list I,int C,list w,list E,list M,dict assign):
    import copy
    cdef:
        dict E_dict = {}
        list nonconf_E = []
        set V = set()
        set H = set()
        dict assign_temp = {}
        set D = set()
        set D_i = set()
    E_dict = E_dict_func(E) #コンフリクトを表す枝を管理する辞書
    nonconf_E = nonconf_E_func(E,I)
    #1
    V = set(I)
    H = set(I)-set(E_dict.keys()) #アイテムの集合V内のコンフリクトの数が0のアイテムの集合をHとし、V=V\Hとする。
    V = V-H #コンフリクトの数が0のアイテムを含まないアイテム集合
    #2
    while V!=set():
        if V==set(): #V=φならば終了
            break
        else:
            D = Johnson(V,set(E)) #DはグラフG内の最大のクリーク
            V = V-D
            #print("クリーク",D)
            #print("残りアイテム",V)
        #3
        for i in D:
            V = V | {i}
            D_i = Johnson(V,set(nonconf_E)) | {i} #V ∪ {i} と　コンフリクトがないアイテム同士の組みを要素にもつ枝集合　から最大クリークを見つける
            assign_temp = FFD_for_conflict(list(D_i),C,w,E_dict,M,assign_temp)
            for index,value in assign_temp.items():
                if i in value:
                    V = V - set(value)
                    for new_key in range(len(assign_temp.keys())):
                        if new_key not in set(assign.keys()): #new_keyがassignに登録されていない場合
                            assign[new_key] = value
                            break
                        elif set(assign[new_key])|(set(value)&D_i)==set(value):#assign_temp[index]: #new_keyはassignに登録されていて、正しい割り当てビンの場合
                            assign[new_key] = list(set(assign[new_key]) | set(value))
                            break   
            assign_temp = copy.deepcopy(assign)
            D = D - {i}
    #4
    assign = FFD(list(H),C,w,M,assign) #Hに対してFFDを行う
    return assign

    
    



#入力, 出力を書き出す階層を指定
import os
path = os.path.abspath(".")

cdef int n = 0
cdef list V = []
cdef int C = 150
cdef list w = []
cdef list E = []
cdef list M = []
cdef int group_split_cost = 1
cdef list r = []
cdef int bin_cost = 5
cdef dict E_dict = {}
cdef list I = []
cdef int n1 = 0
cdef int n2 = 0
cdef list I1 = []
cdef list I2 = []
cdef list I1_E = []
cdef list I2_E = []
cdef dict assign = {}
cdef int penalty_sum = 0
cdef dict penalty_dict = {}
cdef list penalty_list = []

for instance_name in ["w120","w250","w500","w1000"]:
    for instance_number in range(10):
        for density in range(10):
                    
            if instance_number==0:################################
                if instance_name=="w120":

                    file_input = path + "/" + "input" + "/" + str("i"+str(instance_name[1:])) + "_"  + str("n"+str(instance_number)) + "_"  + str("d"+str(int(density*10))) + ".txt" #インスタンスの規模_インスタンスの番号_グラフの密度

                    n = 0
                    V = []
                    C = 150
                    w = []
                    E = []
                    M = []
                    group_split_cost = 1
                    r = []
                    bin_cost = 5

                    with open(file_input,mode="r") as f:
                        for line in f:
                            if "Number of items\t" in line:
                                n = int(line.split("\t")[-2])
                                V = list(range(n))
                            elif "Capacity of bins\t" in line:
                                C = int(line.split("\t")[-2])
                            elif "w\t" in line:
                                w.append(int(line.split("\t")[-1]))
                            elif "e\t" in line:
                                E.append(tuple(map(int,line.split("\t")[-1].rstrip("\n").strip("(").strip(")").split(","))))
                            elif "g\t" in line:
                                M.append(tuple(map(int,line.split("\t")[-1].rstrip("\n").strip("(").strip(")").strip(",").split(","))))
                            elif "Costs of spliting groups\t" in line:
                                r = [int(i) for i in line.split("\t")[-2][1:-1].split(",")]
                            elif "Costs of using bins\t" in line:
                                bin_cost = int(line.split("\t")[-2])


                    #アイテムの分割
                    ##コンフリクトの数が大きい順、大きさが大きい順にアイテムを並べ替える
                    E_dict = {}
                    E_dict = E_dict_func(E)
                    I = sorted(V,key=lambda x:(degree(x,E_dict), w[x]),reverse=True)
                    ##I1とI2の要素の個数を決める
                    n1 = 0
                    n2 = 0
                    n1 = int(np.ceil(n*bin_cost/(bin_cost+r[0])))
                    n2 = n-n1
                    ##I1とI2を定める
                    I1 = []
                    I2 = []
                    I1 = I[:n1]
                    I2 = I[n1:]
                    ##枝集合の更新
                    I1_E = []
                    I2_E = []
                    I1_E = update_E(E,I1)
                    I2_E = update_E(E,I2)
                    
                    for heuristics in range(3,9):
                         if heuristics==8:########################################################################

                            #ヒューリスティクスの選択
                            assign = {}
                            assign = Gendreau(I1,C,w,I1_E,M,assign)
                            
                            if heuristics==3:
                                assign = H3(I2,C,w,I2_E,M,r,assign)
                            elif heuristics==4:
                                assign = H4(I2,C,w,I2_E,M,r,assign)
                            elif heuristics==5:
                                assign = H5(I2,C,w,I2_E,M,r,assign)
                            elif heuristics==6:
                                assign = H6(I2,C,w,I2_E,M,r,assign)
                            elif heuristics==7:
                                assign = H7(I2,C,w,I2_E,M,r,assign)
                            elif heuristics==8:
                                assign = H8(I2,C,w,I2_E,M,r,assign)

                            #結果の出力　→　出力をテキストファイルに書き出す
                            bin_counter=len(assign.keys())

                            penalty_sum = 0
                            penalty_dict = {}
                            penalty_list = []
                            for items in assign.values():
                                penalty_list = list(set(group(item,M) for item in items)) #そのビンを使用しているグループの番号
                                for group_name in penalty_list:
                                    penalty_dict[group_name] = penalty_dict.setdefault(group_name,0) + 1
                            penalty_sum = sum(v*r[i] for i,v in penalty_dict.items())

                            file_output =  path + "/" + "output" + "/" + "Myheuristics_n0" + "/" + str("i"+str(instance_name[1:])) + "_"  + str("n"+str(instance_number)) + "_"  + str("d"+str(int(density*10))) + "_" + str("H"+str(heuristics)) + ".txt" #インスタンスの規模_インスタンスの番号_グラフの密度
                            with open(file_output,mode="w") as f:
                                f.write("Number of Bins\t"+str(bin_counter)+"\n\n")
                                f.write("Bin cost\t"+str(bin_cost)+"\n\n")
                                f.write("Penalty\t"+str(penalty_sum)+"\n\n")
                                f.write("obj\t"+str(penalty_sum+bin_counter*bin_cost))
```
