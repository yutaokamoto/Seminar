# ファイル読み込み 

## バイナリファイルの場合

```python
filename = "/Users/yutaokamoto/Desktop/yutaokamoto_mac/SeminarC125.9.clq.txt"
E=[]
with open(filename,mode="rb") as f:
    lines = f.readlines()
    for i,v in enumerate(lines):
        line = v.decode()
        if "number of vertices" in line:
            n = int(line.split(" ")[-1])
        if line[0]=="e":
            start = i
            break
    E = lines[start:-1] #ファイル名とEOFを除く
    for i in range(len(E)):
        E[i] = tuple(map(int,E[i][2:-1].decode().split(" ")))
```
